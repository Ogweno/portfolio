# Binarization
let’s apply binarization, starting with the easiest part. I want to perform binarization on the CLASS variable. In this case, I want to convert it to the values 0 or 1, exactly the concept of binarization.

The first step to accomplish this is to count the total distinct records, i.e., the total number of categories in the CLASS column. To do this, you will use the SELECT DISTINCTkeyword on the CLASS column.

```python
#SELECT DISTINCT class FROM TB_DATA;

try:
    # Connect to SQLite database
    conn = sqlite3.connect('/content/breast_cancer.db')
    cursor = conn.cursor()

    # Execute the SQL query
    cursor.execute("SELECT DISTINCT class FROM TB_DATA")

    # Fetch the results
    distinct_classes = cursor.fetchall()

    # Print the distinct classes
    for row in distinct_classes:
        print(row[0])  # Assuming class is the first column in the result set

except sqlite3.Error as e:
    print("Error:", e)
finally:
    # Close connection
    conn.close()
```

```python
'''
SELECT
     CASE
	WHEN class = 'no-recurrence-events' THEN 0
	WHEN class = 'recurrence-events'    THEN 1
     END as class
FROM cap03.TB_DATA;
'''

try:
    # Connect to SQLite database
    conn = sqlite3.connect('/content/breast_cancer.db')
    cursor = conn.cursor()

    # Execute the SQL query
    cursor.execute("""
        SELECT
            CASE
                WHEN class = 'no-recurrence-events' THEN 0
                WHEN class = 'recurrence-events' THEN 1
            END as class
        FROM TB_DATA;
    """)

    # Fetch the results
    results = cursor.fetchall()

    # Print the results
    for row in results:
        print(row[0])  # Assuming class is the first column in the result set

except sqlite3.Error as e:
    print("Error:", e)
finally:
    # Close connection
    conn.close()

'''
we use the value 0 for the negative class and the value 1 for the positive class.
You will find this in practically any data science project;
it is something that remains a standard.
'''
```

# Binarization of the “irradiat” variable

```python
# SELECT DISTINCT irradiat FROM TB_DATA
'''
SELECT
     CASE
       WHEN irradiat = 'no'  THEN 0
       WHEN irradiat = 'yes' THEN 1
     END AS irradiat
FROM cap03.TB_DATA;
'''

try:
    # Connect to SQLite database
    conn = sqlite3.connect('/content/breast_cancer.db')
    cursor = conn.cursor()

    # Execute the SQL query
    cursor.execute("""
        SELECT DISTINCT
            CASE
                WHEN irradiat = 'yes' THEN 1
                WHEN irradiat = 'no' THEN 0
            END as irradiat
        FROM TB_DATA;
    """)

    # Fetch the results
    results = cursor.fetchall()

    # Print the results
    for row in results:
        print(row[0])  # Assuming irradiat is the first column in the result set

except sqlite3.Error as e:
    print("Error:", e)
finally:
    # Close connection
    conn.close()

'''
when the value of IRRADIAT is equal to NO, it will be converted to 0,
and when the value of IRRADIAT is YES, it will be converted to 1,
adhering to the convention that negative categories are 0 and positive categories are 1.
'''
```
